# Annual precipitation of QYZ from flux tower data 

```python
qyz3 = pd.read_csv(r"D:\IITM project\validation data\QYZ\FxMt_QYZ_2003_30m_01.csv")
qyz4 = pd.read_csv(r"D:\IITM project\validation data\QYZ\FxMt_QYZ_2004_30m_01.csv")
qyz3.drop(0, axis = 0, inplace = True)
qyz4.drop(0, axis = 0, inplace = True)
qyz3 = qyz3.astype('float32')
qyz4 = qyz4.astype('float32')
```

# Annual precipitation for BNS

```python
bns = pd.read_excel(r"D:/BNS_Rainfall_2010_2014.xlsx", sheet_name=None)
```

```python
bns10 = bns["2010"]
bns11 = bns["2011"]
bns12 = bns["2012"]
bns13 = bns["2013"]
bns14 = bns["2014"]
b = [bns10, bns11, bns12, bns13, bns14]
```

```python
df = []
for d in b:
    d.drop(0, axis = 0, inplace = True)
    df.append(d.astype('float32'))
```

```python
for d in df:
    plt.figure(figsize = (12,8))
    plt.scatter(d.index, d.precipitation)
    plt.show()
```

```python
for d in df:
    if d.iloc[0,0]%4==0:
        print(f"Annual precipitation of BNS_{int(d.iloc[0,0])}: {d.iloc[:,-1].mean()*366*24*2}")
    else:
        print(f"Annual precipitation of BNS_{int(d.iloc[0,0])}: {d.iloc[:,-1].mean()*365*24*2}")
```
